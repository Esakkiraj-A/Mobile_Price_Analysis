```python
train_df=pd.read_csv('train.csv',encoding='latin-1')
train_df.head()
```

```python
train_df.columns
```

```python
columns_to_change = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']
for column in columns_to_change:
   train_df[column] = train_df[column].replace({0: 'No', 1: 'Yes'})
```

```python
train_df.head()
```

```python
train_df['price_range'] = train_df['price_range'].replace({'Low Cost': 1, 'Medium Cost': 2, 'High Cost': 3})
```

```python
train_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for column in columns_to_change:
    train_df[column] = train_df[column].replace({'No': 0, 'Yes': 1})

# Assuming 'train_df' is your dataset
# Calculate the correlation matrix
corr_matrix = train_df.corr()
price_range_corr = corr_matrix['price_range'].sort_values(ascending=False)

# Iterate over variables and create lmplots
for column in price_range_corr.index:
  if column != 'price_range':  # Exclude plotting price_range against itself
    sns.lmplot(x=column, y='price_range', data=train_df)
    plt.title(f'Price Range vs {column}')
    plt.show()
```
